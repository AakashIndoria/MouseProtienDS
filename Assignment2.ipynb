```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
# Lets take the first parameter, criterion:
```

```python
modelOne = DecisionTreeClassifier(criterion='entropy', random_state=1)
modelOne.fit(xTrain,yTrain)
forcast = modelOne.predict(xTest)
accuracy_score(yTest, forcast)
```

```python
modelOne = DecisionTreeClassifier(criterion='gini', random_state=1)
modelOne.fit(xTrain,yTrain)
forcast = modelOne.predict(xTest)
accuracy_score(yTest, forcast)
```

```python
# As seen, the accuracy is greater with 'gini' split. Let's keep it that way.
```

```python
# Let's tune the second parameter: max_depth.
```

```python
# Please note that the CV (cross validation) parameter has been set to 9.
```

```python
# In order to follow a rule of thumb, that cv should be similar to the square root of actual column value, which was 77.
```

```python
depthRange = range(1, 11)
depth_scores = []
for depth in depthRange:
    mdOne = DecisionTreeClassifier(criterion='gini', max_depth=depth, random_state=1)
    d_scores = cross_val_score(mdOne, x, y, cv=9, n_jobs=4, scoring='accuracy')
    depth_scores.append(d_scores.mean())
```

```python
plt.plot(depthRange, depth_scores)
plt.xlabel('Max depth values')
plt.ylabel('Accuracy:')
plt.title('Accuracy difference for different max_depth values:')
```

```python
maxVar = max(depth_scores)
```

```python
optimum_depth = depth_scores.index(maxVar)+1
```

```python
maxVar
```

```python
optimum_depth
```

```python
### So, for best results, the max_depth should be taken as 10. ###
```

```python
### Let's consider the best value for the next parameter: min_samples_split ###
```

```python
sampleRange = range(1, 11)
sample_scores = []
for samples in sampleRange:
    mdTwo = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_split=samples, random_state=1)
    smp_scores = cross_val_score(mdTwo, x, y, cv=9, n_jobs=4, scoring='accuracy')
    sample_scores.append(smp_scores.mean())
```

```python
plt.plot(sampleRange, sample_scores)
plt.xlabel('Min Sample Split values')
plt.ylabel('Accuracy:')
plt.title('Accuracy difference for different min_samples_split values:')
```

```python
### So, let's keep the min_sample_split value as 2. ###
```

```python
# Now let's move on to the next parameter: min_samples_leaf
```

```python
sampleRange = range(0, 5)
leaf_scores = []
for leaf in sampleRange:
    mdThree = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_split=2, min_samples_leaf=leaf, random_state=1)
    lf_scores = cross_val_score(mdThree, x, y, cv=9, n_jobs=4, scoring='accuracy')
    leaf_scores.append(lf_scores.mean())
```

```python
plt.plot(sampleRange, leaf_scores)
plt.xlabel('Min Sample Leaf values')
plt.ylabel('Accuracy:')
plt.title('Accuracy difference for different min_samples_leaf values:')
```

```python
### As seen in the above graph the max accuracy can be achevied by taking the min_samples_leaf value as 1. ###
```

```python
### Let's move on to the next parameter: max_features ###
```

```python
sampleRange = range(1, 10)
feat_scores = []
for feat in sampleRange:
    mdFour = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_split=2, min_samples_leaf=1, max_features=feat, random_state=1)
    ft_scores = cross_val_score(mdFour, x, y, cv=9, n_jobs=4, scoring='accuracy')
    feat_scores.append(ft_scores.mean())
```

```python
plt.plot(sampleRange, feat_scores)
plt.xlabel('max_features value:')
plt.ylabel('Accuracy:')
plt.title('Accuracy difference for different max_features values:')
```

```python
maxFeatVal = max(feat_scores)
```

```python
# As we can see that the accuracy has increased when the value of the max_features=None. #
```

```python
# Let's consider the last value for our tree; max_leaf_nodes #
```

```python
sampleRange = range(1, 10)
mLeaf_scores = []
for mleaf in sampleRange:
    mdFive = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_split=2, min_samples_leaf=1, max_features=None, max_leaf_nodes=mleaf, random_state=1)
    maxLeaf_scores = cross_val_score(mdFive, x, y, cv=9, n_jobs=4, scoring='accuracy')
    mLeaf_scores.append(maxLeaf_scores.mean())
```

```python
plt.plot(sampleRange, feat_scores)
plt.xlabel('max_leaf_nodes value:')
plt.ylabel('Accuracy:')
plt.title('Accuracy difference for different max_leaf_nodes values:')
```
